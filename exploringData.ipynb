# Exploring Data
#### Functions to explore data
- _.to_string()_
- _.head(value)_      #gives the values from the top
- _.tails(value)_     #gives the value from the bottom
- *_.info()_*           #provides the structure of the dataset
- *_data.describe()_*   #it describes the data as shown below
- *.isnull()*           #to check the null value if there is any it will return true

```python
import pandas as pd
data=pd.read_csv('apple_products.csv')
print(data.to_string())
```
